## Define value table $Q(s, a)$

#### Create the $Q(s, a)$ table

## Define the policy $\pi(s)$

#### Create the policy $\pi(s)$

```python
def policy(state, epsilon=0.):
    if np.random.random() < epsilon:
        return np.random.randint(4)
    else:
        av = action_values[state]
        return np.random.choice(np.flatnonzero(av == av.max())) 
```

#### Test the policy with state (0, 0)

```python
action = policy((0,0))
print(f"Action taken in state (0,0): {action}")
```

```python
plot_policy(action_values, frame)
```

## Implement the algorithm


</br></br></br>




<div style="text-align:center">
    Adapted from Barto & Sutton: "Reinforcement Learning: An Introduction".
</div>

```python
def on_policy_mc_control(policy, action_values, episodes, gamma = 0.99, epsilon = 0.2, alpha = 0.2):#on policy: same policy to exploration and learning processes, action_values= Q_values
    #alpha: speed  at which we push the estimate in the direction of the new returns
    sa_returns = {} #store returns for specific state , action pair
    for episode in range (1, episodes+1):
        state = env.reset()
        done = False # if task is over 
        transitions = [] #Historical data: state, action, reward for each step
        
        while not done: # loop executed until the end of the episode
            action = policy(state, epsilon)
            next_state, reward, done,_ = env.step(action) # make the agent take the action and face consequences
            transitions.append([state, action, reward])
            state = next_state
            
        G= 0
        for state_t, action_t, reward_t in reversed(transitions): #backward to calculate return 
            G= reward_t + gamma * G
            qsa = action_values[state_t][action_t]
            action_values[state_t][action_t] += alpha * (G-qsa)
```

```python
on_policy_mc_control(policy, action_values, episodes = 10000)
```
